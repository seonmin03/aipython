```python
#bar()
import matplotlib.pyplot as plt
import pandas as pd
df=pd.DataFrame({'도시':['서울', '부산', '인천', '대구'], '인구 수': [9904312, 3448737, 2890451, 2466052]})
plt.rc('font', family='NanumGothic')
plt.bar(df['도시'], df['인구 수'])
plt.title('도시별 인구 수')
plt.xlabel('도시')
plt.ylabel('인구 수')
```

```python
#mission1
import pandas as pd
df=pd.read_excel('data1.xlsx')
plt.plot(df['name'], df['math'],'r', marker='v', label='수학점수')
plt.plot(df['name'], df['kor'], 'g--', marker='o', label='국어점수')
plt.title('성적 그래프')
plt.xlabel('이름')
plt.ylabel('점수')
plt.legend()
plt.show()
```

```python
#seaborn_scatterplot()
import seaborn as sns
iris_data=pd.read_csv('iris.csv')
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_data)
plt.title('Scatter plot by seabron', fontsize=20)
plt.show()
```

```python
#seaborn_countplot()
data_1 = pd.DataFrame({'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]})
sns.countplot(x='category', hue='category', data=data_1)
plt.show
```

```python
#mission3
data_penguin=pd.read_csv('penguins.csv') #sns.load_dataset('penguins'): seaborn 자체에 이 파일이 있기 때문에 사용할 수 있음
sns.barplot(x='species', y='body_mass_g', data=data_penguin, hue='sex')
plt.show()
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm',data=data_penguin, hue='species')
plt.show()
sns.scatterplot(x='bill_length_mm', y='body_mass_g',data=data_penguin, hue='species')
plt.show()
sns.regplot(x='bill_length_mm', y='bill_depth_mm', data=data_penguin)
plt.show()
```
