```python
import pandas as pd
df=pd.read_csv('iris.csv')
df.info()
df.describe()
```

```python
df['species'].value_counts()
df.isna().sum()
df[df.duplicated()]
idx=(df['sepal_length']==5.8)&(df['petal_width']==1.9)
df.loc[idx,:]
df2=df.drop_duplicates()
```

```python
print(df2.groupby('species').sum())
print(df2.groupby('species').mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.bar(df2['species'], df2['petal_length'])
plt.title('Sepal Length Visualization')
plt.xlabel('Data')
plt.show()
sns.barplot(x='species', y='petal_length', data=df2)
```

```python
plt.hist(x='sepal_length', bins=4, data=df2)
plt.show()
plt.hist(x='sepal_width', bins=3, data=df2)
plt.show()
plt.hist(x='petal_length', bins=5, data=df2)
plt.show()
plt.hist(x='petal_width', bins=3, data=df2)
plt.show()
```

```python
plt.scatter(df2['sepal_length'], df2['petal_length'])
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df2)
```
